```python
# Import necessary libraries for OCR and image handling
import pytesseract
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
# Upload a file from local machine
uploaded = files.upload()
```

```python
# Process each uploaded image using OCR
for filename in uploaded.keys():
    image = cv2.imread(filename)  # Load the image
    if image is not None:
        cv2_imshow(image)  # Display the image
        extracted_text = pytesseract.image_to_string(image)  # Extract text
        print("Text recognized from image:\n")
        print(extracted_text)
    else:
        print(f"Failed to read image: {filename}. Please check the file format.")
```

```python
from google.colab import files
from pdf2image import convert_from_path
import pytesseract
from PIL import Image

uploaded = files.upload()
pdf_path = next(iter(uploaded))
pages = convert_from_path(pdf_path)
for i, page in enumerate(pages):
    text = pytesseract.image_to_string(page)
    print(text)
```

## Observations

- OCR accuracy depends heavily on image quality and resolution.
- Preprocessing like resizing and grayscale conversion improves results.
- OCR works better on printed text compared to handwritten text.
- Multi-page PDF text extraction is slower compared to image files.
